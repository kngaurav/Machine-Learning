**Import libraries**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.metrics import RocCurveDisplay
```

**Importing the dataset**

```python
df = pd.read_csv("Social_Network_Ads.csv")
```

```python
display(df)
```

```python
df.info()
```

**Splitting the dataset into the Training set and Test set**

```python
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
X = df.drop(columns=['Purchased','User ID'])
Y = df['Purchased']
display(X)
```

```python
display(Y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

**Feature Scaling**

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

**Fitting SVM to the Training set**

```python
clf = svm.SVC(kernel='rbf',gamma='scale')
clf.fit(x_train,y_train)
```

**Predicting the Test set results**

```python
y_pred=clf.predict(x_test)
y_pred
```

**Making the Confusion Matrix**

```python
confusion_matrix(y_test,y_pred)
pd.crosstab(y_test,y_pred,rownames = ['Actual'],colnames=['Predicted'],margins=True)
```

**Visualising the Training set results**

```python
training_score = clf.score(x_train,y_train)*100
print("Train data score is :",training_score)
```

```python
RocCurveDisplay.from_estimator(clf,x_train,y_train)
```

**Visualising the Test set results**

```python
test_score=accuracy_score(y_test,y_pred)*100
print("Test data score is :",test_score)
```

```python
RocCurveDisplay.from_estimator(clf,x_test,y_test)
```
